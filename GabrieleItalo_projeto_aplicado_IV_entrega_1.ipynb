# **Introdução**

O monitoramento de dados históricos é essencial para o progresso das sociedades modernas. A análise de séries temporais permite compreender padrões em saúde, economia e clima. Através desse estudo, é possível antecipar crises e otimizar recursos públicos (MORETTIN; TOLOI, 2018).

Segundo Box, Jenkins e Reinsel (2015), a previsão é vital para a redução de incertezas futuras. Dessa forma, a ciência de dados cumpre um papel transformador na gestão da informação global.

A análise de séries temporais é também crucial para o planejamento em cadeias de suprimentos. No Brasil, a distribuição nacional de produtos exige previsões precisas devido à dimensão territorial. Projeções assertivas permitem otimizar estoques e reduzir custos logísticos (CHOPRA; MEINDL, 2016). A eficiência operacional depende diretamente da qualidade das estimativas de demanda regional.

Além da eficiência econômica, a precisão analítica promove a produção e o consumo sustentáveis. Previsões acuradas evitam o excesso de estoques e o consequente desperdício de recursos essenciais. Essa abordagem alinha-se ao Objetivo de Desenvolvimento Sustentável (ODS) 12 da ONU (ONU, 2015).
A sustentabilidade nas cadeias de suprimentos é, portanto, um reflexo da maturidade tecnológica.

Falhas na estimativa de demanda resultam em prejuízos severos para a rede de distribuição. Erros de previsão comprometem a disponibilidade de produtos e elevam gastos operacionais (HYNDMAN; ATHANASOPOULOS, 2018).
Portanto, a escolha de um método robusto de previsão é uma necessidade crítica e atual.

A literatura aponta que não existe um modelo universalmente superior para todos os cenários (WOLPERT; MACREADY, 1997). Diferentes fenômenos exigem abordagens distintas para capturar suas especificidades. Essa premissa justifica a comparação entre diferentes abordagens para o mesmo fenômeno.

Este trabalho aborda dois paradigmas distintos para a modelagem de séries temporais. O intuito é comparar o desempenho de abordagens estatísticas e de aprendizado de máquina. Essa dualidade metodológica visa identificar a técnica mais eficaz para o contexto analisado. Os resultados contribuirão para o aprimoramento das previsões de distribuição no setor.
